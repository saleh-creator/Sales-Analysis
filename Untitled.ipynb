### Merging 12 monthes of sales data into a single file

```python
files = [file for file in os.listdir("./Sales_Data")]

all_monthes_data=pd.DataFrame()


for file in files:
    df=pd.read_csv("./Sales_Data/"+file)
    all_monthes_data=pd.concat([all_monthes_data,df])
    
all_monthes_data.head()
```

## Q4: What products are most sold together?
### we can use apriori algorithm but we can explore the data first

```python
all_data.head()
```

```python
all_data.shape
```

```python
all_data['Order ID'].value_counts()
```

### it seems we have some orders with the same IDs ,It means that there are orders that have been sold together

```python
df = all_data[all_data["Order ID"].duplicated(keep=False)]

df['Grouped'] = df.groupby(["Order ID"])["Product"].transform(lambda x:','.join(x))

df = df[['Order ID','Grouped']].drop_duplicates()
```

```python
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()


for row in df["Grouped"]:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    

count.most_common(10)

```
